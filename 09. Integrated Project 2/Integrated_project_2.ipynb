**Описание данных:**
В данных содержатся параметры добычи и очистки золота из руды. Данные индексируются датой и временем получения информации (признак `date`).

Данные находятся в трёх файлах:
`gold_industry_train.csv` — обучающая выборка;
`gold_industry_test.csv` — тестовая выборка;
`gold_industry_full.csv` — исходные данные.

В распоряжении сырые данные: их просто выгрузили из хранилища. Некоторые параметры недоступны, потому что замеряются и/или рассчитываются значительно позже. Из-за этого в тестовой выборке отсутствуют некоторые признаки, которые могут быть в обучающей. Также в тестовом наборе нет целевых признаков.

Исходный датасет содержит обучающую и тестовую выборки со всеми признаками.

**Технологический процесс:**
* Rougher feed — исходное сырье
* Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
* Xanthate — ксантогенат (промотер, или активатор флотации);
* Sulphate — сульфат (на данном производстве сульфид натрия);
* Depressant — депрессант (силикат натрия).
* Rougher process (англ. «грубый процесс») — флотация
* Rougher tails — отвальные хвосты
* Float banks — флотационная установка
* Cleaner process — очистка
* Rougher Au — черновой концентрат золота
* Final Au — финальный концентрат золота

Параметры этапов
* air amount — объём воздуха
* fluid levels — уровень жидкости
* feed size — размер гранул сырья
* feed rate — скорость подачи

**Наименование признаков:**
Наименование признаков должно быть такое:
[этап].[тип_параметра].[название_параметра]
Пример: `rougher.input.feed_ag`

Возможные значения для блока [этап]:
* rougher — флотация
* primary_cleaner — первичная очистка
* secondary_cleaner — вторичная очистка
* final — финальные характеристики

Возможные значения для блока [тип_параметра]:
* input — параметры сырья
* output — параметры продукта
* state — параметры, характеризующие текущее состояние этапа
* calculation — расчётные характеристики

**Целевые признаки:**
* `rougher.output.recovery` — эффективность обогащения чернового концентрата
* `final.output.recovery` — эффективность обогащения финального концентрата

### Проверка корректности расчета эффективности обогащения

Проверим, что эффективность обогащения в `rougher.output.recovery` рассчитана правильно на обучающей выборке. Для этого посчитаем ее по формуле и оценим MAE с заданными значениями в этом столбце:

```python
# напишем функцию, рассчитывающую эффективность обогащения по формуле
def calculate_recovery(c, f, t):
    return c * (f - t) / (f * (c - t)) * 100
```

```python
# применим эту функцию на обучающей выборке
# доля золота в концентрате после флотации/очистки
c = data_train['rougher.output.concentrate_au']
# доля золота в сырье/концентрате до флотации/очистки
f = data_train['rougher.input.feed_au']
# доля золота в отвальных хвостах после флотации/очистки
t = data_train['rougher.output.tail_au']

recovery_train = calculate_recovery(c, f, t)

# вычислим MAE между нашими расчетами и значением признака
print('MAE', mean_absolute_error(recovery_train, data_train['rougher.output.recovery']))
```

MAE получилось ничтожно маленькой, значит эффективность обогащения рассчитана правильно на обучающей выборке.

Это один из целевых признаков, значит, если в нем есть пропуски, мы можем заполнить их по формуле на тренировочной выборке:

```python
print('Количество пропусков в rougher.output.recovery', data_train['rougher.output.recovery'].isna().sum())
```

Пропусков нет в этом столбце, значит, не нужно их заполнять

## Модель

### Вычисление итоговой sMAPE

Напишем функцию, вычисляющую метрику качества sMAPE - Symmetric Mean Absolute Percentage Error:

```python
def calculate_smape(target, prediction):
    return 1 / len(target) * np.sum(abs(target - prediction) / ((abs(target) + abs(prediction)) * 0.5)) * 100
```

```python
def calculate_final_smape(smape_rougher, smape_final):
    return 0.25 * smape_rougher + 0.75 * smape_final
```

### Подготовка данных для обучения моделей

Подготовим тренировочные и тестовые данные для обучения моделей. Для этого нужно удалить из тренировочной выборки признаки, относящиеся к финальным характеристикам, которых нет в тестовой выборке. В качестве целевого признака будем рассматривать оба требуемых целевых признака: `rougher.output.recovery`, `final.output.recovery`

```python
features_train_final = data_train.drop(column_difference, axis=1)
target_train_final = data_train['final.output.recovery']
```

Для тестовых данных найдем целевые признаки в полном датафрейме по индексу:

```python
features_test_final = data_test
targets_test = data_full[data_full.index.isin(data_test.index)][['rougher.output.recovery', 'final.output.recovery']]
target_test_final = targets_test['final.output.recovery']
```

Для вычисления sMAPE на этапе флотации подготовим признаки и целевой признак, содержащий только характеристики, относящиеся к этому этапу, на тренировочной и тестовой выборках:

```python
features_train_rougher = features_train_final.filter(regex='rougher')
target_train_rougher = data_train['rougher.output.recovery']
features_train_rougher.columns
```

```python
features_test_rougher = features_test_final.filter(regex='rougher')
target_test_rougher = targets_test['rougher.output.recovery']
```

Во время обзора данных мы заметили, что диапазоны возможных значений разных признаков отличаются в несколько десятков раз, поэтому далее проведем масштабирование данных, чтобы все признаки воспринимались как одинаково важные:

```python
# масштабируем признаки для данных на этапе флотации 
scaler = StandardScaler()
scaler.fit(features_train_rougher)
features_train_rougher = scaler.transform(features_train_rougher)
features_test_rougher = scaler.transform(features_test_rougher) 

# масштабируем признаки для данных на остальных этапах 

numeric_columns = features_train_final.columns.values.tolist()
numeric_columns.remove(numeric_columns[0])

scaler.fit(features_train_final[numeric_columns])
features_train_final[numeric_columns] = scaler.transform(features_train_final[numeric_columns])
features_test_final[numeric_columns] = scaler.transform(features_test_final[numeric_columns]) 
```

### Обучение и оценка моделей

Реализуем функцию, которая с помощью GridSearchCV подбираем лучшую модель среди переданных параметров. Модели оцениваются по sMAPE кросс-валидацией:

```python
def find_best_model_smape(model, parameters):
    grid_rougher = GridSearchCV(model, parameters, scoring=smape_scorer, cv=5, n_jobs=-1)
    grid_rougher.fit(features_train_rougher, target_train_rougher)

    print('Наилучшая модель на этапе флотации', grid_rougher.best_params_)
    print('sMAPE наилучшей модели на этапе флотации', abs(grid_rougher.best_score_))
    
    grid_final = GridSearchCV(model, parameters, scoring=smape_scorer, cv=5, n_jobs=-1)
    grid_final.fit(features_train_final, target_train_final)

    print('Наилучшая модель после двух очисток', grid_final.best_params_)
    print('sMAPE наилучшей модели после двух очисток', abs(grid_final.best_score_))
    
    best_smape_train = calculate_final_smape(abs(grid_rougher.best_score_), abs(grid_final.best_score_))
    print('Итоговый sMAPE на тренировочной выборке', best_smape_train)
    
    return grid_rougher.best_estimator_, grid_final.best_estimator_
```

#### Регрессия дерева решений

Воспользуемся DecisionTreeRegressor для поиска наилучшей модели случайного дерева по метрике sMAPE:

```python
%%time
dtr_model = DecisionTreeRegressor(random_state=12345)
dtr_parameters = {'max_depth': range(1,8)}

smape_scorer = make_scorer(calculate_smape, greater_is_better=False)

dtr_best_rougher_estimator, dtr_best_final_estimator = find_best_model_smape(dtr_model, dtr_parameters)
```

#### Регрессия случайного леса

```python
%%time
rfr_model = RandomForestRegressor(random_state=12345)
rfr_parameters = {"n_estimators":[50, 100], 'max_depth': [2, 5]}

rfr_best_rougher_estimator, rfr_best_final_estimator = find_best_model_smape(rfr_model, rfr_parameters)
```

#### Линейная регрессия

```python
%%time
lr_model = LinearRegression()
lr_parameters = {'fit_intercept' : [True, False], 'normalize' : [True, False]}

lr_best_rougher_estimator, lr_best_final_estimator = find_best_model_smape(lr_model, lr_parameters)
```

### Тестирование и проверка на адекватность лучшей модели

#### Тестирование лучшей модели

Проверим, как лучшая модель регрессии случайного леса отрабатывает на тестовой выборке:

```python
# определим функцию, вычиляющую sMAPE модели на тестовой выборке
def test_model(best_rougher_estimator, best_final_estimator):
    predictions_test_rougher = best_rougher_estimator.predict(features_test_rougher)
    predictions_test_final = best_final_estimator.predict(features_test_final)

    best_smape_test = calculate_final_smape(calculate_smape(target_test_rougher, predictions_test_rougher),
                                            calculate_smape(target_test_final, predictions_test_final))
    print('Итоговый sMAPE на тестовой выборке', best_smape_test) 
```
